```python
def remove_out_sqft(df):
    new_data = pd.DataFrame()
    for loc, loc_df in df.groupby('location'):
        loc_mean = np.mean(loc_df['price_per_sqft'])
        loc_std = np.std(loc_df['price_per_sqft'])
        filtered_sqft = loc_df[((loc_df['price_per_sqft'])>=(loc_mean - loc_std)) & ((loc_df['price_per_sqft'])<=(loc_mean + loc_std))]
        new_data = pd.concat([new_data,filtered_sqft], ignore_index=True)
    return new_data
```

```python
def remove_out_sqft_bhk(df):
    new_data = pd.DataFrame()
    for loc, loc_df in df.groupby('location'):
        loc_mean = np.mean(loc_df['sqft_per_bedroom'])
        loc_std = (np.std(loc_df['sqft_per_bedroom']))*3
        filtered_sqft = loc_df[((loc_df['sqft_per_bedroom'])>=(loc_mean - loc_std)) & ((loc_df['sqft_per_bedroom'])<=(loc_mean + loc_std))]
        new_data = pd.concat([new_data,filtered_sqft], ignore_index=True)
    return new_data
```

```python
data = remove_out_sqft(data)
```

```python
data = remove_out_sqft_bhk(data)
```

```python
def remove_price_out(df):
    price_out_ar = np.array([])
    for loc, loc_df in df.groupby('location'):
        loc_dict = {}
        for bhk, bhk_df in loc_df.groupby('BHK'):
            loc_dict[bhk] = {
                'mean': np.mean(bhk_df['price_per_sqft']),
                'count' : bhk_df.shape[0]
            }
        for bhk, bhk_df in loc_df.groupby('BHK'):
            stats = loc_dict.get(bhk-1)
            if stats and stats['count']>5:
                price_out = bhk_df[(bhk_df['price_per_sqft']<stats['mean'])].index.values
                price_out_ar = np.append(price_out_ar, price_out)
    return df.drop(price_out_ar, axis = 'index')
```

```python
data = remove_price_out(data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import r2_score
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
X = data.drop(columns = ['price'])
y = data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=172)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse = False), ['location']), remainder = 'passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_lr)
```

```python
import pickle
```

```python
pickle.dump(pipe,open('LR_Bangalore_Model.pkl','wb'))
```
